```python
# text : the text of the tweet 
# not include stopwords
from itertools import chain
from collections import Counter
from nltk import ngrams
from nltk.corpus import stopwords
import itertools

def text_analysis(df):
    num_print = 100

    # length of tweet
    len_tweets = df.text.apply(lambda x:len(x.split()))
    print('length of tweets:\n{}\n'.format(len_tweets.describe()))

    # word count
    set_of_stopwords = set(stopwords.words('english'))
    corpus_include_stopwords = Counter(chain.from_iterable([[word for word in text.split() if word not in set_of_stopwords] for text in df.text.tolist()]))
    corpus = Counter(chain.from_iterable([text.split() for text in df.text.tolist()]))

    print('word count include stop words: \n{}\n'.format(corpus_include_stopwords.most_common(num_print)))
    print('word count without stop words: \n{}\n'.format(corpus.most_common(num_print)))

    # bigram count
    n = 2
    corpus_for_ngram = Counter(
        chain.from_iterable(
            [['-'.join(sorted(ngram)) for ngram in ngrams(text.split(), n)] for text in df.text.tolist()]
        )
    )
    print('bigram count: \n{}\n'.format(corpus_for_ngram.most_common(num_print)))

    # biterm count

    get_biterm = lambda x: [sorted(biterm) for biterm in itertools.combinations(x, 2) if biterm[0] != biterm[1]]
    get_biterm_without_stopwords = lambda x: [
        sorted(biterm) for biterm in itertools.combinations(x, 2) if biterm[0] != biterm[1] and biterm[0] not in set_of_stopwords and biterm[1] not in set_of_stopwords
    ]

    corpus_for_biterm = Counter(
        chain.from_iterable(
            [['-'.join(biterm) for biterm in get_biterm(text.split())] for text in df.text.tolist()]
        )
    )
    print('biterm count: \n{}\n'.format(corpus_for_biterm.most_common(num_print)))

    corpus_for_biterm_without_stopwords = Counter(
        chain.from_iterable(
            [['-'.join(biterm) for biterm in get_biterm_without_stopwords(text.split())] for text in df.text.tolist()]
        )
    )
    print('biterm count without stopwords: \n{}\n'.format(corpus_for_biterm_without_stopwords.most_common(num_print)))
    
text_analysis(df)
```

# Cross sectional analysis

```python
# target : this denotes whether a tweet is about a real disaster (1) or not (0)
df.target.value_counts()
```

```python
# keyword 
def get_crosstab(column):
    num_print = 30
    crosstab = pd.crosstab(df[column], df['target'])
    crosstab['sum'] = crosstab.sum(axis=1)
    return crosstab.sort_values('sum', ascending=False)[:num_print]

get_crosstab('keyword')
```

```python
# location
get_crosstab('location')
```

```python
# not disaster
text_analysis(df.loc[df.target==0])
```

```python
# real disaster
text_analysis(df.loc[df.target==1])
```
