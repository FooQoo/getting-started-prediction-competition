```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df_1 = pd.read_csv('../fact/valid_cv_merge_1.csv')
df_2 = pd.read_csv('../fact/valid_cv_merge_2.csv')
df_3 = pd.read_csv('../fact/valid_cv_merge_3.csv')
df_4 = pd.read_csv('../fact/valid_cv_merge_4.csv')
df_5 = pd.read_csv('../fact/valid_cv_merge_5.csv')

df = pd.concat([df_1, df_2, df_3, df_4, df_5]).reset_index(drop=True)
```

```python
sns.distplot(df[df['target']==0].text, kde=True, rug=False)
sns.distplot(df[df['target']==1].text, kde=True, rug=False)
```

```python
sns.distplot(df[df['target']==0].category, kde=True, rug=False)
sns.distplot(df[df['target']==1].category, kde=True, rug=False)
```

```python
sns.jointplot(df[df.target==0].text, df[df.target==0].category, kind="kde" )
```

```python
sns.jointplot(df[df.target==1].text, df[df.target==1].category, kind="kde" )
```
